```python
# Perfect sensor
def sense(x):
    return x
```

```python
def simulate(Δt, x, u):
    x += Δt * u
    return x
```

```python
def PControl(t, x, K, xd):
    return K * (xd - x)
```

```python
def control(t, x):
    return PControl(t, x, 0.5, np.array([2., 10.]))
```

```python
import numpy as np

tf = 10            # Final time
Δt = 0.01          # Time step
time = np.linspace(0., tf, int(tf / Δt) + 1)  # Time interval

# Initial conditions
x = array([4, 5.])
x_log = [np.copy(x)]
position = [np.copy(x)]

for t in time:
    y = sense(x)
    u = control(t, y)    
    x = simulate(Δt, x, u)
    x_log.append(np.copy(x))
    if t%1 == 0: # Mark position every second in a 10-seconds interval
        position.append(np.copy(x))
    
position = np.array(position)
x_log = np.array(x_log)
```

```python
import numpy as np
%matplotlib inline

ax = gca()
ax.set_aspect('equal', adjustable='box')
grid()
plot(x_log[:,0], x_log[:,1])
plot(position[:,0], position[:,1], 'o', color='red')
arrow(x_log[0,0], x_log[0,1], (x_log[1,0] - x_log[0,0])/0.05, (x_log[1,1] - x_log[0,1])/0.05, 
      shape='full', lw=4, length_includes_head=True, head_width=.05, color='r')
```

```python
import numpy as np

initial = np.array([4, 5.])
final = np.array([2, 10.])

tf = 10         # Final time
Δt = 0.01      # Time step
time = np.linspace(0., tf, int(tf / Δt) + 1)  # Time interval

x = copy(initial)
x_log = [np.copy(x)]

for t in time:
    y = sense(x)
    u = PControl(t, x, 0.5, np.array([2., 10.])) #k = 0.5
    x = simulate(Δt, x, u)
    x_log.append(np.copy(x))
    
x_log = np.array(x_log)
#adding a fixed time increment to the previous time 
#appending the new time value to the end of the array
time = np.append(time, float(time[-1]) + Δt)

plot(time, x_log[:,1], label='Y(t)')
plt.axhline(y = 10, color = 'r', linestyle = '--')
plot(time, x_log[:,0], label='X(t)')
plt.axhline(y = 2, color = 'r', linestyle = '--')
plt.title("X & Y Position over Time")
legend()
show()

error = np.array([(final - x) for x in x_log])
plot(time, error[:,1], label='Y Error')
plot(time, error[:,0], label='X Error')
plt.axhline(y = 0, color = 'r', linestyle = '--')
plt.title("X & Y Error over Time")
legend()
show()
```
